# **Childcare Cost**

## **Executive Summary**

The cost of childcare in the United States has been a constant challenge for many American families. Understanding the many features that play a role into how much families can afford, can be beneficial for policy makers, employers, and families when planning ahead. The stakeholders for our project can even include the childcare facilities, based on the demographics of the area they are located in. Business owners would be able to see the average that families can afford.

With the recent change of leadership in New York city, there has been many conversations about affordable childcare. An article from Bloomberg discussed the different challenges families face given the high cost of childcare and the upcoming trial of 2-Care. This will be able to provide care for about 2,000 two-year-olds in high need areas, in hopes that it will be able to expand across the city in four years.

Similarly, in another high cost of living city across the country we are seeing a similar option for families. The city of San Francisco is expanding their free and subsidized childcare to families earning under \$230,000/year and subsidized costs for those making up to \$310,000. Based on the data collection from Children's Council of San Francisco shows that full-time care can cost nearly \$30,000 a year. This will be financed through a commercial rent tax which was approved in 2018, which has not been spent according to Bloomberg.

Seeing two major cities offer free or subsidized childcare is the start of change for many American families. Through our project we can assess the cost for other cities across the country, leading to understanding what changes can be made to help childcare become more affordable.

## **EDA**

Reviewing the initial correlations that may exist between child care cost and variables that may explain it.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting relevant columns for the pairplot
selected_columns = [
    'households','pr_p','fme_2018','mme_2018',
    'mcsa','mfccsa'
]

sns.pairplot(child_care_cost[selected_columns])
plt.suptitle('Pairplot of Childcare Costs', y=1.02) # Adjust suptitle to prevent overlap
plt.show()
```

```python
variables_to_plot = [
    'mcsa','mfccsa',

]

# Determine grid size
n_vars = len(variables_to_plot)
n_cols = 2  # For example, 3 columns
n_rows = (n_vars + n_cols - 1) // n_cols # Calculate rows needed

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten() # Flatten the 2D array of axes for easy iteration

for i, var in enumerate(variables_to_plot):
    sns.histplot(child_care_cost[var], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequency')

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Now exploring how costs vary across states, before going into county-level depth.

```python
# Merging the data sets on county_fips_code
cost_comp = pd.merge(child_care_cost, county, on='county_fips_code')
```

```python
# Calculate the mean mfccsa for each state
state_mcsa = cost_comp.groupby('state_name')['mcsa'].mean().reset_index()

# Sort the states by mfccsa for better visualization
state_mcsa_sorted = state_mcsa.sort_values(by='mcsa', ascending=False)

# Create the bar plot with a single color
plt.figure(figsize=(15, 12)) # Adjust figure size for wide and long plot
sns.barplot(x='mcsa', y='state_name', data=state_mcsa_sorted, color='#A2E5B2', legend=False)

# Add labels to the data points, aligned to the right inside the bar
for i, row in enumerate(state_mcsa_sorted.itertuples()): # Data point labeling
    # Place the label slightly inside the bar, and use black color for contrast
    plt.text(row.mcsa - 2, i, f'{row.mcsa:.2f}', color='black', ha="right", va="center")

plt.title('Average Weekly Center-Based Care Cost (School Age) by State')
plt.xlabel('Average Weekly MCSA')
plt.ylabel('State Name')
plt.tight_layout()
plt.show()
```

```python
# Calculate the mean mfccsa for each state
state_mfccsa = cost_comp.groupby('state_name')['mfccsa'].mean().reset_index()

# Sort the states by mfccsa for better visualization
state_mfccsa_sorted = state_mfccsa.sort_values(by='mfccsa', ascending=False)

# Create the bar plot with a single color
plt.figure(figsize=(15, 12)) # Adjust figure size for wide and long plot
sns.barplot(x='mfccsa', y='state_name', data=state_mfccsa_sorted, color='#9FDBE6', legend=False)

# Add labels to the data points, aligned to the right inside the bar
for i, row in enumerate(state_mfccsa_sorted.itertuples()): # Data point labeling
    # Place the label slightly inside the bar, and use black color for contrast
    plt.text(row.mfccsa - 2, i, f'{row.mfccsa:.2f}', color='black', ha="right", va="center")

plt.title('Average Weekly Family Childcare Cost (School Age) by State')
plt.xlabel('Average Weekly MFCCSA')
plt.ylabel('State Name')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import pandas as pd

index_label = ['h_under6_both_work','h_under6_f_work','h_under6_m_work','h_under6_single_m']
aggregated_six = child_care_cost.groupby('study_year')[['h_under6_both_work','h_under6_f_work','h_under6_m_work','h_under6_single_m']].sum()
value_raw = aggregated_six.mean()

# Create a Series to easily sort both labels and values
sorted_data = pd.Series(value_raw, index=index_label).sort_values(ascending=True)
under_six_care = sorted_data.index
value = sorted_data.values

plt.figure(figsize=(12, 6)) # Make the graph wider
bars = plt.barh(under_six_care, value, color='pink') # Use barh for horizontal bars

# Define a custom formatter for compact millions
def millions_formatter(x, pos):
    return f'{x/1000000:.0f}M'

# Apply the formatter to the x-axis
formatter = mticker.FuncFormatter(millions_formatter)
plt.gca().xaxis.set_major_formatter(formatter)

# Add data labels
for bar in bars:
    plt.text(bar.get_width() + 100000, bar.get_y() + bar.get_height()/2,
             millions_formatter(bar.get_width(), None),
             va='center', ha='left', color='black')

plt.xlabel("Sum of Households") # Swap x and y labels
plt.ylabel("Household Type") # New label for the y-axis
plt.title("Households with Children Under 6 - Yearly avg")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
